GET Http Method
The GET method is used to retrieve information from the given server using a given URL. 
The GET method sends the encoded user information appended to the page request. 
The page and the encoded information are separated by the ‘?’ character

```python
import requests

# Make a GET request to the GitHub API to retrieve information about a user
username = 'octocat'  # Example username
url = f'https://api.github.com/users/harish-AK'
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Print the response content (the data retrieved)
    user_data = response.json()
    print(user_data)
else:
    # Print an error message if the request was unsuccessful
    print('Error:', response.status_code)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import time
import pytz

# Function to scrape data and print it
def scrape_data():
    for _ in range(10):
        try:
            # Get current time
            tz_NY = pytz.timezone('Asia/Kolkata')
            datetime_NY = datetime.now(tz_NY)
            current_time = datetime_NY.strftime("%H:%M:%S - (%d/%m)")
            print(f'At time : {current_time} IST')

            # Make request to Yahoo Finance
            response = requests.get('https://finance.yahoo.com/crypto/?fr=sycsrp_catchall')
            text = response.text
            html_data = BeautifulSoup(text, 'html.parser')

            # Extract headings
            headings = html_data.find_all('tr')[0]
            headings_list = [x.text for x in headings][:10]

            # Extract data for top 5 cryptocurrencies
            data = []
            for x in range(1, 6):
                row = html_data.find_all('tr')[x]
                column_value = row.find_all('td')
                coin_data = {}
                for i in range(10):
                    coin_data[headings_list[i]] = column_value[i].text
                data.append(coin_data)

            # Print data
            for coin in data:
                print()
                print(coin)
                print()

        except Exception as e:
            print("Error:", e)

        # Wait for 6 seconds before scraping again
        time.sleep(2)

# Run the scraping function for 10 iterations
scrape_data()
```

# Retrieve paragh

```python
import requests
from bs4 import BeautifulSoup

# URL of the website to scrape
url = "https://www.geeksforgeeks.org/chat-bot-in-python-with-chatterbot-module/"

# Send a GET request to the website
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find the paragraph element
    paragraph_element = soup.find("p")
    
    # Extract the text of the paragraph
    if paragraph_element:
        paragraph_text = paragraph_element.get_text()
        print(paragraph_text)
    else:
        print("No paragraph element found on the page.")
else:
    print("Failed to retrieve the webpage.")
```
